```python
%%time

TOIs = [datetime.datetime(2025, 10, 9, 14, 0), datetime.datetime(2025, 10, 9, 14, 0), datetime.datetime(2025, 11, 8, 15, 0),
        datetime.datetime(2025, 8, 17, 15, 0), datetime.datetime(2025, 5, 4, 15, 0), datetime.datetime(2025, 9, 21, 15, 0),
        datetime.datetime(2025, 9, 30, 15, 0), datetime.datetime(2025, 9, 16, 15, 0), datetime.datetime(2025, 7, 27, 15, 0)]

HARPNUMs = [13866, 13876, 14003, 13641, 13171, 13777, 13845, 13773, 13561]

for ii in range(2, len(HARPNUMs)):
    TOI = TOIs[ii]
    HARPNUM = HARPNUMs[ii]

    print(f"{HARPNUM}_{TOI}")

    T_REC_i = TOI.strftime('%Y-%m-%dT%H:%M:%S')
    T_REC_f = (TOI + datetime.timedelta(seconds = 60 * 5)).strftime('%Y-%m-%dT%H:%M:%S')

    path_to_directory = os.path.join(main_dir, f"data/{HARPNUM}")

    if os.path.isdir(path_to_directory):
        print('Directory exists!\n')

    else:
        print('Making directory...')
        os.mkdir(path_to_directory)
        print('Directory formed!\n')


    res1 = Fido.search(a.Time(T_REC_i, T_REC_f),
                      a.jsoc.Series('hmi.sharp_cea_720s'),
                      a.jsoc.PrimeKey('HARPNUM', HARPNUM),
                      a.jsoc.Segment('Br') & a.jsoc.Segment('magnetogram') &
                      a.jsoc.Segment('continuum') &
                      a.jsoc.Segment('bitmap') & a.jsoc.Segment('conf_disambig'),
                      a.jsoc.Notify('pj22rs046@iiserkol.ac.in'))

    print(res1)

    if len(res1) >0 :
        Fido.fetch(res1, path = path_to_directory)
```
